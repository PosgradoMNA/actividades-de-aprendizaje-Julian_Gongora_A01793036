#6 Escalar los datos con la funcin: 

Los datos obtenidos contienen caractersticas de varias dimensiones y escalas en total. Las diferentes escalas de las caractersticas de los datos afectan negativamente al modelado de un conjunto de datos. Las varianzas de los valores sin ninguna transformacion o normalizacion son muy distintas y puede afectar a la tecnica PCA ya que utiliza las varianzas

```python

# Procedemos a realizar una Estandarizacin: cambia la escala de un conjunto de datos para que tenga una media de 0 y una desviacin estndar de 1
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled = scaler.fit_transform(df_num)
scaled_df = pd.DataFrame(scaled, columns=df_num.columns)
scaled_df.head()
```

**Verificacin:** Observamos que todas las varianzas quedan en 1 despues de la transformacion. Vamos a tomar 3 columnas para la verificacin: X1, X2 y X12

```python
cols =['X1','X5','X12']
```

```python
summary = {
"Varianzas": scaled_df[cols].var(),
"Valor Min": scaled_df[cols].min(),
"Valor Max": scaled_df[cols].max()
}
pd.DataFrame(summary, index = cols).transpose()
```
